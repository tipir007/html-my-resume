```python
# Connecting Google Colab with Google BigQuery

from google.colab import auth
from google.cloud import bigquery
auth.authenticate_user()
print('Authenticated')
```

```python
project_id_akun = 'hacktive-christopher' # Gunakan GoolgeCloudPlatform Project-ID Anda
client = bigquery.Client(project=project_id_akun)
```

```python
# Show the dataset

df = client.query('''

select *
from bigquery-public-data.san_francisco_bikeshare.bikeshare_trips
limit 10

''').to_dataframe()

df
```

```python
# No. 1 Jumlah total perjalanan dari setiap stasiun selama tahun 2023. urut dari yg terbanyak

df1 = client.query('''

SELECT start_station_name,
COUNT (start_station_id) as total_perjalanan
FROM bigquery-public-data.san_francisco_bikeshare.bikeshare_trips
WHERE EXTRACT(YEAR FROM start_date) = 2023
GROUP BY start_station_name
ORDER BY total_perjalanan DESC


''').to_dataframe()

df1
```

```python
# No 2 Rata-rata durasi perjalanan (dalam menit) yang dilakukan oleh pelanggan yang menggunakan e-bikes (num_ebikes_available > 0), Juni 2023
df2 = client.query( '''
select AVG(duration_sec)/60 as rata_rata_durasi,
COUNT (bike_number) as total_ebike

from bigquery-public-data.san_francisco_bikeshare.bikeshare_trips
where EXTRACT(MONTH FROM start_date) = 6 AND EXTRACT (YEAR FROM start_date) = 2023
''' ). to_dataframe()

df2
```

```python
# No 3 Stasiun dengan ketersediaan sepeda paling sedikit pada akhir bulan Desember 2023 (show table )
df0 = client.query( '''
SELECT * FROM bigquery-public-data.san_francisco_bikeshare.bikeshare_station_status
LIMIT 50

 ''' ).to_dataframe()
df0
```

```python
# No 3 Stasiun dengan ketersediaan sepeda paling sedikit pada akhir bulan Desember 2023
df3 = client.query('''
SELECT station_id, num_bikes_available
FROM bigquery-public-data.san_francisco_bikeshare.bikeshare_station_status
WHERE EXTRACT(MONTH FROM TIMESTAMP_MILLIS(last_reported * 1000)) = 12
  AND EXTRACT(YEAR FROM TIMESTAMP_MILLIS(last_reported * 1000)) = 2023
ORDER BY num_bikes_available ASC
LIMIT 1

''').to_dataframe()
df3
```

```python
# No 4 Jumlah total perjalanan oleh pengguna wanita pada Sem 1 2024
df4 = client.query('''
SELECT COUNT(trip_id) as total_perjalanan
FROM bigquery-public-data.san_francisco_bikeshare.bikeshare_trips
WHERE EXTRACT(YEAR FROM start_date) = 2024
AND EXTRACT(MONTH FROM start_date) <= 6
AND member_gender = 'female'

''').to_dataframe()
df4

```

```python
# No 5 URUTAN daerah dengan jumlah stasiun terbanyak hingga paling sedikit
df00 = client.query( '''
SELECT bikeshare_station_info.region_id, bikeshare_station_info.station_id, bikeshare_station_info.name
FROM bigquery-public-data.san_francisco_bikeshare.bikeshare_station_info
LIMIT 50

 ''' ).to_dataframe()
df00
```

```python
# No 5 URUTAN daerah dengan jumlah stasiun terbanyak hingga paling sedikit
df5 = client.query( '''
SELECT bikeshare_station_info.region_id, COUNT(bikeshare_station_info.station_id) as total_stasiun
FROM bigquery-public-data.san_francisco_bikeshare.bikeshare_station_info
GROUP BY bikeshare_station_info.region_id
ORDER BY total_stasiun DESC

 ''' ).to_dataframe()
df5
```

```python
# NO 6 Pada jam berapa perjalanan terbanyak dilakukan dalam satu hari selama bulan mei 2025

df6 = client.query('''
SELECT EXTRACT(HOUR FROM start_date) as jam, COUNT(trip_id) as total_perjalanan
from bigquery-public-data.san_francisco_bikeshare.bikeshare_trips
WHERE EXTRACT(MONTH FROM start_date) =  5 AND EXTRACT(YEAR FROM start_date) = 2025
GROUP BY jam
ORDER BY total_perjalanan DESC
LIMIT 1
''').to_dataframe()
df6
```
